```python
import pandas as pd
import sqlite3
```

```python
 # connect to existing database
conn = sqlite3.connect('Dem_table.db')
cur = conn.cursor()
```

```python

cur.execute('''CREATE TABLE Pandemic(
  L_ID INT NOT NULL,
  T_ID INT NOT NULL,
  V_ID INT NOT NULL,
  Current_Cases INT NOT NULL,
  New_Cases INT NOT NULL,
  Recovered INT NOT NULL,
  Deaths INT NOT NULL,
PRIMARY KEY (L_ID, T_ID, V_ID),
FOREIGN KEY (L_ID) REFERENCES Location(L_ID),
FOREIGN KEY (T_ID) REFERENCES Time(T_ID),
FOREIGN KEY (V_ID) REFERENCES Virus(V_ID));''')
```

```python
cur.execute('''CREATE TABLE Vaccination(
  Vac_ID INT NOT NULL,
  T_ID INT NOT NULL,
  L_ID INT NOT NULL,
  Doses_Received INT NOT NULL,
  First_Dose INT NOT NULL,
  Second_Dose INT NOT NULL,
  Third_Dose INT NOT NULL,
PRIMARY KEY (Vac_ID, T_ID, L_ID),
FOREIGN KEY (L_ID) REFERENCES Location(L_ID),
FOREIGN KEY (T_ID) REFERENCES Time(T_ID),
FOREIGN KEY (Vac_ID) REFERENCES Vaccine(Vac_ID));''')
```

```python
cur.execute('''CREATE TABLE ICU(
  L_ID INT NOT NULL,
  T_ID INT NOT NULL,
  Indicator VARCHAR(30) NOT NULL,
  Source VARCHAR(30) NOT NULL,
  ICU_Cases INT NOT NULL,
PRIMARY KEY (L_ID, T_ID),
FOREIGN KEY (L_ID) REFERENCES Location(L_ID),
FOREIGN KEY (T_ID) REFERENCES Time(T_ID));''')
```

```python
cur.execute('''CREATE TABLE Virus(
V_ID INT PRIMARY KEY NOT NULL,
Name VARCHAR(30) NOT NULL);''')
```

```python
cur.execute('''CREATE TABLE Vaccine(
Vac_ID INT PRIMARY KEY NOT NULL,
Name VARCHAR(30) NOT NULL);''')
```

```python
cur.execute('''CREATE TABLE Location(
L_ID INT PRIMARY KEY NOT NULL,
Country VARCHAR(30) NOT NULL,
Region VARCHAR(30) NOT NULL,
Population INT NOT NULL);''')
```

```python
cur.execute('''CREATE TABLE Time(
T_ID INT PRIMARY KEY NOT NULL,
Month INT NOT NULL,
Quarter INT NOT NULL,
Year INT NOT NULL);''')
```

```python
sql_query = """
SELECT L.Region,
       T.Month,
       SUM(P."CURRENT_CASES") AS "Total Cases",
       SUM(P.DEATHS) AS "Total Deaths",
       I.Source,
       SUM(I."ICU_CASES") AS "Total ICU Cases",
       V.Name AS "Vaccine Name",
       SUM(Va."DOSES_RECEIVED") AS "Total Doses Received"
FROM Pandemic P
JOIN Location L ON P."L_ID" = L."L_ID"
JOIN Time T ON P."T_ID" = T."T_ID"
JOIN ICU I ON I."L_ID" = L."L_ID" AND I."T_ID" = T."T_ID"
JOIN Vaccination Va ON Va."L_ID" = L."L_ID" AND Va."T_ID" = T."T_ID"
JOIN Vaccine V ON Va."Vac_ID" = V."Vac_ID"
GROUP BY L.Region,
         T.Month,
         I.Source,
         V.Name
Limit 5;
"""

df = pd.read_sql_query(sql_query, conn)
df
```
